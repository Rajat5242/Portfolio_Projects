```python
from textblob import TextBlob
```

```python
polarity=[]
for text in df['Text']:
    try:
        polarity.append(TextBlob(text).sentiment.polarity)
    except:
        polarity.append(0)
```

```python
polarity[0:10]
```

```python
df['polarity']=polarity
```
